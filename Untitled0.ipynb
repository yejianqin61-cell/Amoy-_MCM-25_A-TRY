```python
pip install ultralytics


!pip install easyocr


!pip install opencv-python

import os
import cv2
import math
from datetime import datetime

from ultralytics import YOLO
import easyocr
model = YOLO("best.pt")

reader = easyocr.Reader(['en'], gpu=False)

ENTRY_TIME = datetime.strptime(
    "2025-01-01 10:15:00",
    "%Y-%m-%d %H:%M:%S"
)

EXIT_TIME = datetime.strptime(
    "2025-01-01 13:40:00",
    "%Y-%m-%d %H:%M:%S"
)

HOURLY_RATE = 2.0

def calculate_parking_fee(entry_time, exit_time, hourly_rate):
    """
    Calculate parking fee based on entry and exit time.

    Rules:
    - Duration is calculated in hours
    - Any fraction of an hour is rounded up
    - Fee = charged_hours × hourly_rate
    """


    duration_seconds = (exit_time - entry_time).total_seconds()


    duration_hours = duration_seconds / 3600

    charged_hours = math.ceil(duration_hours)


    total_fee = charged_hours * hourly_rate

    return charged_hours, total_fee

IMAGE_DIR = "test_images"
OUTPUT_DIR = "output_results"


os.makedirs(OUTPUT_DIR, exist_ok=True)

os.makedirs(IMAGE_DIR, exist_ok=True)


for img_name in os.listdir(IMAGE_DIR):

    img_path = os.path.join(IMAGE_DIR, img_name)
    image = cv2.imread(img_path)


    if image is None:
        continue

    results = model(image)[0]


    if results.boxes is None or len(results.boxes) == 0:
        print(f"No plate detected in {img_name}")
        continue

    confidences = results.boxes.conf.cpu().numpy()
    best_index = confidences.argmax()
    best_box = results.boxes[best_index]


    x1, y1, x2, y2 = map(int, best_box.xyxy[0])


    h, w, _ = image.shape
    x1 = max(0, x1)
    y1 = max(0, y1)
    x2 = min(w, x2)
    y2 = min(h, y2)

    plate_crop = image[y1:y2, x1:x2]



    ocr_results = reader.readtext(plate_crop)

    if len(ocr_results) > 0:

        plate_text = max(ocr_results, key=lambda x: x[2])[1]
    else:
        plate_text = "UNREADABLE"


    charged_hours, total_fee = calculate_parking_fee(
        ENTRY_TIME,
        EXIT_TIME,
        HOURLY_RATE
    )


    cv2.rectangle(
        image,
        (x1, y1),
        (x2, y2),
        (0, 255, 0),
        2
    )


    cv2.putText(
        image,
        f"Plate: {plate_text}",
        (20, 30),
        cv2.FONT_HERSHEY_SIMPLEX,
        0.8,
        (0, 255, 0),
        2
    )

    cv2.putText(
        image,
        f"Fee: RM {total_fee:.2f}",
        (20, 65),
        cv2.FONT_HERSHEY_SIMPLEX,
        0.8,
        (0, 255, 0),
        2
    )

    output_path = os.path.join(OUTPUT_DIR, img_name)
    cv2.imwrite(output_path, image)


    print(
        f"{img_name} | Plate: {plate_text} | "
        f"Hours: {charged_hours} | Fee: RM {total_fee:.2f}"
    )
```
